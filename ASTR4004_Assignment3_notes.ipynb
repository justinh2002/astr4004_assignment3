Task 3: The radial metallicity relation in simulated data

radial_metallicity_fit_and_residuals.png

2 panel figure: – (a) Logarithmic density plot of RGal. vs. A(O), with a linear fit and legend.
– (b) Residuals of the fit, RGal. vs. ∆A(O)

**Linear Fit**: The fitted equation is given by $A(O) = -0.033 R_{Gal} + 9.122$.
The negative slope indicates that the gas-phase metallicity $A(O)$ decreases with increasing galactocentric radius. The linear trend generally captures the overall negative gradient of the gas phase metallicity very well, which is to be expected for radial metallicity gradients in disc galaxies, where metallicity decreases outward due to phenomena like inside-out formation and metal enrichment in the central regions.

**RMSE:** is 0.088 which is very low given that the metallicity ranges from roughly 7.8 - 9.2, implying there is a relatively small standard deviation. This suggests that the linear model captures the general trend in gas-phase metallicity very well. I also considered RMSEs for regions of larger residuals (how I set this threshold was to select the regions in which we have residuals greater than one stddev), THEN RMSE for regions with larger residuals become 0.182, which is significantly larger and suggests the need for a more complex model for those specific parts of the galaxy, where a simple linear fit does not suffice. See below on where exactly these regions are located.

**Residual Analysis:** However, limitations in the fit quality at different radii lie in the residuals (deviations from the linear fit). There is a rather systematic trend, with residuals clustering around 0 at lower $R_{Gal}$ and increasing in scatter with increasing radius. The spread in residuals become far more pronounced at larger radii, e.g., $R_{Gal} > 10 \text{kpc}$). The linear model cannot capture all variability in the data, especially in the outer regions here, which is perhaps to be expected for a linear fit. The larger residuals in the outer regions imply that the metallicity could possibly follow a more complex functional form than a linear relation, perhaps one that flattens beyond a certain radius, where a transition to a less well-mixed or accretion-dominated regime may have occurred. Other effects such as radial migration, as mentioned earlier, can affect the radial metallicity distribution in the outer regions as well.

**Bin Size**: I chose this bin size to balance resolution and data coverage. With fewer bins, important local features would be smoothed out, so we may lose valuable information metallicity variation. More bins comes at a cost of noise production, and the statistics per bin would become unreliable due to lower particle counts. Therefore, a larger number of bins could help identify smaller-scale features like spiral arms, but this would come at the cost of increased noise and unreliable median values in regions with fewer data points. This is a fine-tuning problem.

Overall, the linear model provides a good first-order approximation of the radial metallicity gradient but is unable to capture all the underlying complexities of the gas-phase radial metallicity distribution. As we saw earlier, this is particularly obvious in the outer regions where the gradient might flatten or vary due effects that cannot be captured through a linear model, e.g., radial migration and gas accretion. Further to this, if we examine the residual patterns, we find that they exhibit non-random/non-stochastic patterns, indicating that the metallicity distribution is affected by more than just radial distance, which we thought earlier. This suggests a need for higher-order or more complex modeling, potentially incorporating spiral arms (see also that error is magnified in those regions) or localised enrichment processes.
